```python
from CointegrationLiveTester import CointegrationLiveTester


data = {
    
    'live_test_start_date': '2022-5-19',
    'filename': 'IndustrialsLowerThreshold',
    'max_backtests': 5,
    'max_allowable_allocation': 0.7,
    'initial_cash': 4000,
    'description': 'This is the first real live trade that we are starting on 2022-05-18',
    'print_component_summary': True,
}

lt = CointegrationLiveTester(data)

lt.run(print_backtest_summary=True)
lt.plot_combined_strategies()
lt.perform_t_test(alpha=0.05, verbose=True)
lt.perform_monte_carlo(total_trading_days=60, simulations=100, plot_results=True)
lt.print_summary()
```

```python

start_equity = lt.os_data['total'][0]
start_equity

end_equity = lt.os_data['total'][-1]
end_equity

profit = end_equity - start_equity
profit

total_return = profit / start_equity
total_return

total_days = (lt.os_data.iloc[-1].name - lt.os_data.iloc[0].name).days
total_days

daily_return = total_return / total_days
daily_return

annualized_return = ((1 + daily_return)**365) - 1
annualized_return
```
